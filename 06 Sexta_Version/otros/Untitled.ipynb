```python
import mysql.connector
import tkinter as tk

# Crear una nueva ventana
window_ = tk.Tk()
window_.minsize(width=140, height=200)
window_.title("Ingresar datos")

base_sql = tk.StringVar()
tabla_sql = tk.StringVar()

# Crear etiquetas y campos de entrada
label_user = tk.Label(window_, text="Usuario:")
label_user.grid(row=1, padx=10, pady=5)
field_user = tk.Entry(window_)
field_user.grid(row=2, padx=10, pady=5)

label_password = tk.Label(window_, text="Contraseña:")
label_password.grid(row=3, padx=10, pady=5)
field_password = tk.Entry(window_, show="*")
field_password.grid(row=4, padx=10, pady=5)

def get_bases():
    """
    Funcion para obtener las bases disponibles
    """
    global user
    global password
    global base_radios

    user = field_user.get()
    password = field_password.get()

    config = {
        'user': user,
        'password': password,
        'host': '132.248.182.174',
        'port': 3306
    }

    conn = mysql.connector.connect(**config)
    cursor = conn.cursor()
    query = "SHOW DATABASES"

    cursor.execute(query)
    # Obtener los resultados
    results = cursor.fetchall()

    bases = []
    for i in results:
        if i[0][0:5] == "BSRN_":
            bases.append(i[0])

    # Eliminar los RadioButtons anteriores (si los hay)
    if 'base_radios' in globals():
        for radio in base_radios:
            radio.destroy()

    base_radios = []
    for option, base in enumerate(bases):
        radio = tk.Radiobutton(
            window_,
            text=base,
            value=base,
            variable=base_sql,
            command=lambda: get_tables(conn, cursor, len(bases)),
            padx=5,
            pady=5
        )
        radio.grid(column=2, row=option, pady=5)
        base_radios.append(radio)

    return conn, cursor

def get_tables(conn, cursor, integer):
    tables = []
    base = base_sql.get()
    query = "USE " + base
    cursor.execute(query)
    query = "SHOW tables"
    cursor.execute(query)
    # Obtener los resultados
    results = cursor.fetchall()
    for i in results:
        if i[0][0:5] == "bsrn_":
            tables.append(i[0])

    # Eliminar los RadioButtons anteriores de tablas (si los hay)
    if hasattr(window_, 'tabla_radios'):
        for radio in window_.tabla_radios:
            radio.destroy()

    window_.tabla_radios = []
    for option, table in enumerate(tables):
        radio = tk.Radiobutton(
            window_,
            text=table,
            value=table,
            variable=tabla_sql,
            padx=5,
            pady=5
        )
        radio.grid(column=3, row=option, pady=5)
        window_.tabla_radios.append(radio)

# Crear el botón "Conectarse"
button_accept = tk.Button(window_, text="Conectarse", command=get_bases)
button_accept.grid(row=5, padx=10, pady=5)

window_.mainloop()
```
